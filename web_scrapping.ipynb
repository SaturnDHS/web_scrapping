```python
from bs4 import BeautifulSoup
import urllib.request as ur
import pandas as pd
import re

#ler a página em um formato de texto
urlToScrape = "https://www.worldometers.info/coronavirus"
r = ur.urlopen(urlToScrape).read()
soup = BeautifulSoup(r, "html.parser")
```

```python
#encontrar a tabela com dados de covid
#table = soup.find('table', {'class':'table'})
table = soup.find('table',{'id': 'main_table_countries_today'})
```

```python
#para cada linha da tabela pegar a informação contida
table_rows = table.find_all('tr')
```

```python
#as coolunas do dataframe serão o header da tabela na página
headers = [header.text.strip() for header in table.find_all('th')]
```

```python
#para os dados pegaremosa a informação de cada linha-vvalor e utilizaremos o re para retirar espaços vazios e afins
data = []
for tr in table_rows:
    td = tr.find_all('td')
    #row = [i.text for i in td]
    row = [re.sub('\s+', ' ', i.text).strip() for i in td]
    data.append(row)
```

```python
#dataframe concluído, a primeira linha e primeira coluna sobraram, estamos prontos para utilizar a informação
df = pd.DataFrame(data, columns=headers)
df = df.iloc[1:,1:]
```
